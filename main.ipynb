Uniqueness analysis

```python
perfect_dupes = df[df.duplicated()]
perfect_dupes_count = len(perfect_dupes)

print(f"Perfect deplicate rows: {perfect_dupes_count}")
print(f"Percentage of total dataset: {(perfect_dupes_count / len(df)) * 100:.2f}%")
```

```python
# Define logical business keys
business_keys = ['Order ID','Product ID', 'Customer ID','Order Date']

# Find groups where the business key appears more than once
biz_dupe_groups = df.groupby(business_keys).size().reset_index(name='count')
biz_dupe_groups = biz_dupe_groups[biz_dupe_groups['count'] > 1]

print(f"Number of duplicated business key groups: {len(biz_dupe_groups)}")
```

```python
if len(biz_dupe_groups) > 0:
    # count total records involved in the duplications
    total_involved = biz_dupe_groups['count'].sum()
    print(f"Total records involved in business key duplicates: {total_involved}")
    
    sample_key = biz_dupe_groups.iloc[0][business_keys].tolist()
    sample_rows = df[(df[business_keys[0]] == sample_key[0]) & (df[business_keys[1]] == sample_key[1]) & (df[business_keys[2]] == sample_key[2])]
    print("\nSample duplicated group:")
    print(sample_rows.to_string())
    
else:
    print("No business key duplicates found. Unique dimension : Pass (Green)")
```

Profile the nature of the duplicates

```python
biz_dups = df[df.duplicated(subset=business_keys,keep=False)]

def count_column_conflicts(groups_df,key_cols,compare_cols):
    conflicts = {}
    for col in compare_cols:
        if col in key_cols:
            continue
        conflict_count = groups_df.groupby(key_cols)[col].nunique().gt(1).sum()
        conflicts[col] = conflict_count
    return conflicts

compare_columns = ['Sales', 'Profit', 'Discount', 'Quantity', 'Ship Date', 'Ship Mode', 'Region', 'Category', 'Sub-Category', 'Product Name']

conflicts = count_column_conflicts(biz_dups,business_keys,compare_columns)

print("Columns with conflicts in duplicated business keys groups:")
for col , count in sorted(conflicts.items(),key=lambda x : x[1],reverse=True):
    if count > 0:
        print(f" {col} : {count} groups have conflicting values")
```

Ship Mode:
  Same Day: Delivery within 0-1 days of the order
  First Class: Delivery expected within 1-2 days
  Second Class: Delivey expected within 2-3 days
  Standard Class: Delivery expected with 4-7 days

```python
if 'Shipping_Delay_Days' not in df.columns:
    df["Shipping_Delay_Days"] = (df["Ship Date"] - df["Order Date"]).dt.days
    
def check_timeliness_anomaly(row):
    delay = row['Shipping_Delay_Days']
    mode = row['Ship Mode']
    
    if pd.isnull(delay) or pd.isnull(mode):
        return 'Cannot Assess -  Missing Data'
    if mode == "Same Day" and delay > 1:
        return f'Anomaly: Same Day shipped in {delay} days'
    elif mode == "First Class" and delay > 2:
        return f"Anomaly: First Class shipped in {delay} days"
    elif mode == "Second Class" and delay > 3:
        return f"Anomaly: Second Class shipped in {delay} days"
    elif mode == "Standard Class" and delay > 7:
        return f"Anomaly: Standard Class shipped in {delay} days"
    elif delay < 0:
        return 'Critical Error: Ship Date Before Order Date'
    else:
        return 'Within Expected Window'
df['timeliness_check'] = df.apply(check_timeliness_anomaly, axis=1)

print('Timeliness Check Summary by Ship Mode')
print(df.groupby("Ship Mode")['timeliness_check'].value_counts())
print(f"\nTotal Records with Critical Errors (negative delay):{len(df[df["Shipping_Delay_Days"] < 0])}")
```

```python
anomalities = df[df['timeliness_check'] != 'Within Expected Window']
print(f'Total records with timeliness anomalies: {len(anomalities)}')

if len(anomalities) > 0 :
    print("\nSample of anomalous records:")
    print(anomalities[['Order ID','Ship Mode','Order Date','Ship Date','Shipping_Delay_Days','timeliness_check']].head(10).to_string())
```

```python
overlap = set(df1['Order ID']).intersection(set(df2['Order ID']))
print(f'Overlapping Order IDs between files: {len(overlap)}')

if len(overlap) > 0:
    print("Sample overlapping Order IDs:")
    print(list(overlap)[:10])
    print("\nWARNING: The two source files contain shared orders. This may indicate a data split error or partial duplication at the source level.")
else:
    print("No overlapping Order IDs. The two filed reprsent distinct, non-overlapping data sources.")
```

```python
overlap_id = set(df1['Order ID']).intersection(set(df2['Order ID'])) 

overlap_df1 = df1[df1['Order ID'].isin(overlap_id)]
overlap_df2 = df2[df2['Order ID'].isin(overlap_id)]

print(f"Overlapping Order IDs: {len(overlap_id)}")
print(f"Total records involved: {len(overlap_df1) + len(overlap_df2)}")

sample_id = list(overlap_id)[0]
sample_from_df1 = df1[df1['Order ID'] == sample_id]
sample_from_df2 = df2[df2['Order ID'] == sample_id]

print(f"\nSample Order ID: {sample_id}")
print("\nSample from File 1: ")
print(sample_from_df1.to_string())
print("\nSample From File 2: ")
print(sample_from_df2.to_string())
```

```python
all_overlaps_df1 = df1[df1['Order ID'].isin(overlap_id)].copy()
all_overlaps_df2 = df2[df2['Order ID'].isin(overlap_id)].copy()

all_overlaps_df1['Source_File'] = "Superstore_Orders_q1.csv"
all_overlaps_df2['Source_File'] = "Superstore_Orders_q2.csv"

combined_overlaps = pd.concat([all_overlaps_df1,all_overlaps_df2], ignore_index=True)
combined_overlaps = combined_overlaps.sort_values(['Order ID', 'Source_File'])
```

```python
conflict_summary = {}
for col in df1.columns:
    if col == 'Source_File':
        continue
    conflict_count = combined_overlaps.groupby('Order ID')[col].nunique().gt(1).sum()
    if conflict_count > 0:
        conflict_summary[col] = conflict_count

if conflict_summary:
    print("Column with conflicting value between File1 and File2:")
    for col, count in sorted(conflict_summary.items(),key=lambda x: x[1], reverse=True):
        print(f" {col}:{count} Order IDs have different values")
else:
    print("All overlapping records are identical between file. They are perfect duplicates at the source level.")
```
